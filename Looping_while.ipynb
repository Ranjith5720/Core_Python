### **While Loop :** ( Post defined )
**Syntax :**  
**While** ( key word ) **Condition : logic**

```python
a = 0
while a <= 10:
  print(a)
  a += 1
```

```python
b = int(input("Enter a number :"))
while b % 2 == 0:
  print(b, "is Even number")
else:
  print(b, "is Odd number")
```

```python
a = [2,3,5,8,12,43,65,6,3.2] # list
i = 0
while i < len(a):
  val_i = a[i]
  if val_i % 2 == 0:
    print(a[i],"is even")
  else:
    print(a[i],"is odd")
  i += 1
```

```python
b = (2,5,7,3,90,4,3,2,14) # tuple
i = 0
while i < len(b):
  val_i = b[i]
  if val_i % 2 == 0:
    print(b[i],"is even")
  else:
    print(b[i],"is odd")
  i += 1
```

```python
c = {12,98,23,12,34,55,22,4,2,8,6} # set
i = 0
j = list(c) # convert set to list.
while i < len(j):
  val_j = j[i]
  if val_j % 2 == 0:
    print(j[i],"is even")
  else:
    print(j[i],"is odd")
  i += 1
```

```python
d = {'a':1,'b':2,'c':34,'d':89,'e':2} # dict
i = 0
j = list(d.values()) # convert dict (values) to list.
while i < len(j):
  val_j = j[i]
  if val_j % 2 == 0:
    print(j[i],"is even")
  else:
    print(j[i],"is odd")
  i += 1
```
